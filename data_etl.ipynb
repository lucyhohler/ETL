### Store 2nd CSV into DataFrame

```python
csv_file = "Resources/PovStatsCountry.csv"
income_group_df = pd.read_csv(csv_file)
income_group_df.head()
```

### Clean DataFrame

```python
new_income_group_df = income_group_df[['Country Code', 'Short Name', 'Long Name', 'Currency Unit', 'Region', 'Income Group', 'Source of most recent Income and expenditure data']].copy()
new_income_group_df.head()
```

```python
# Rename columns
df = new_income_group_df.rename(columns={'Country Code': 'code', 'Short Name': 'name', 
                                    'Long Name': 'official_name', 'Currency Unit': 'currency_unit', 'Region': 'region',
                                   'Income Group': 'income_group',
                                    'Source of most recent Income and expenditure data': 'source'})
df.head()
```
